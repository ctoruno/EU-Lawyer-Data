```python
# Libraries needed
import time
import re
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# store root url without page number
url =  "https://dsserron.gr/meli/"
data = requests.get(url).text
soup = BeautifulSoup(data, 'html.parser')
```

```python
print('Classes of each table:')
for table in soup.find_all('table'):
    print(table.get('class'))
```

```python
# Creating list with all tables
tables = soup.find_all('table')
table = soup.find('table', class_='tablepress-id-2')
```

```python
# Defining of the dataframe
df = pd.DataFrame(columns=['Name', 'Address', 'Phone', 'Mobile', 'Fax', 'Email', 'Translations'])
```

```python
# Collecting Ddata
for row in table.tbody.find_all('tr'):    
    # Find all data for each column
    columns = row.find_all('td')
    
    if(columns != []):
        name = columns[0].text.strip()
        address = columns[1].text.strip()
        phone = columns[2].text.strip()
        mobile = columns[3].text.strip()
        fax = columns[4].text.strip()
        email = columns[5].text.strip()
        translations = columns[6].text.strip()
        
        df = df.append({'Name': name,  'Address': address, 'Phone': phone, 
                        'Mobile': mobile, 'Fax': fax, 'Email': email, 
                       'Translations': translations}, ignore_index=True)
```
