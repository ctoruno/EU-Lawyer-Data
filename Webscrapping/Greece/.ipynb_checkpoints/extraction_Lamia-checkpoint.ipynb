```python
# Libraries needed
import time
import re
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# store root url without page number
url =  "https://ds-lamias.gr/?page_id=137"
data = requests.get(url).text
soup = BeautifulSoup(data, 'html.parser')
```

```python
print('Classes of each table:')
for table in soup.find_all('table'):
    print(table.get('class'))
```

```python
# Creating list with all tables
tables = soup.find_all('table')
table = soup.find('table', class_='tablepress tablepress-id-7')
```

```python
# Defining of the dataframe
df = pd.DataFrame(columns=['A_M', 'Adjective', 'Name', 'Father', 'Stable', 'Mobile', 'Email'])
```

```python
# Collecting Ddata
for row in table.tbody.find_all('tr'):    
    # Find all data for each column
    columns = row.find_all('td')
    
    if(columns != []):
        am = columns[0].text.strip()
        adjective = columns[1].text.strip()
        name = columns[2].text.strip()
        father = columns[3].text.strip()
        stable = columns[4].text.strip()
        mobile = columns[5].text.strip()
        email = columns[6].text.strip()
        
        df = df.append({'A_M': am,  'Adjective': adjective, 'Name': name, 
                        'Father': father, 'Stable': stable, 'Mobile': mobile, 
                       'Email': email}, ignore_index=True)
```
